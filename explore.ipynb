```python
df = pd.read_csv("Merged_2025-03-11.csv")
df = pd.DataFrame(df)
df.head()
```

```python
# Filter dataset for "Chicken Breast" in the Name column (case-insensitive)
chicken_breast_df = df[df["Name"].str.contains("chicken breast", case=False, na=False)]

# Display the unique product names and their respective stores
chicken_breast_names = chicken_breast_df[["Name", "Store_Name"]].drop_duplicates()
```

```python
cheapest_chicken_breast = (
    chicken_breast_df.groupby("Store_Name")
    .apply(lambda x: x.nsmallest(10, "Price"))
    .reset_index(drop=True)
)
```
